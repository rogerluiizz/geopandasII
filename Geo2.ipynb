```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
AREA_VCA = gpd.read_file(r'C:\Users\Samsung\Desktop\Geopandas_parteII\VCA.shp')
```

```python
AREA_IMOVEL = gpd.read_file(r'C:\Users\Samsung\Desktop\Geopandas_parteII\Area_Imovel.shp')
```

```python
HIDROGRAFIA = gpd.read_file(r'C:\Users\Samsung\Desktop\Geopandas_parteII\Hidrografia.shp')
```

```python
APP_VCA = gpd.read_file(r'C:\Users\Samsung\Desktop\Geopandas_parteII\APP_VCA.shp')
```

```python
fig, ax = plt.subplots(1,1, figsize=(20, 10))
AREA_VCA.plot(ax=ax, color = 'none', edgecolor = 'black', linewidth = 2)
HIDROGRAFIA.plot(ax=ax, color='blue',edgecolor = 'blue', linewidth = 2)
AREA_IMOVEL.plot(ax=ax, color = 'none', edgecolor = 'red', linewidth = 0.2)
APP_VCA.plot(ax=ax, color='yellow', edgecolor = 'yellow', linewidth = 1)
```

```python
Intersect = gpd.overlay(AREA_VCA, AREA_IMOVEL, how='intersection')
intersect2 = gpd.overlay(AREA_VCA, APP_VCA, how='intersection')
intersect3 = gpd.overlay(AREA_VCA, HIDROGRAFIA, how='intersection')
```

```python
fig, ax = plt.subplots(1,1, figsize=(20, 10))
AREA_VCA.plot(ax=ax, color = 'none', edgecolor = 'black', linewidth = 2)
Intersect.plot(ax=ax, color = 'none', edgecolor = 'red', linewidth = 0.2)
intersect2.plot(ax=ax, color='yellow', edgecolor = 'yellow', linewidth = 1)
intersect3.plot(ax=ax, color='blue', edgecolor = 'blue', linewidth = 2)
```

```python
union = gpd.overlay(APP_VCA, HIDROGRAFIA, how='union')
fig, ax = plt.subplots(1,1, figsize=(20, 10))
AREA_VCA.plot(ax=ax, color = 'none', edgecolor = 'black', linewidth = 2)
union.plot(ax=ax, color = 'none', edgecolor = 'green', linewidth = 1)
```
